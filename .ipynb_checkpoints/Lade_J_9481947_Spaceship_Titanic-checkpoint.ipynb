# 6. Preprocessing
- Im Notebook Data Prep wurden alle missing values entfernt sowie einige features erstellt, um die Genauigkeit der Modelle zu verbessern

```python
# Load data from Data Preparation
X = pd.read_csv('X.csv')
X_test = pd.read_csv('X_test.csv')
y = train['Transported'].copy().astype(int)
```

## Überflüssige features entfernen

```python
# Drop qualitative/redundant/collinear/high cardinality features
X.drop(['Unnamed: 0','PassengerId', 'Group', 'Group_size', 'Age_group', 'Cabin_number'], axis=1, inplace=True)
X_test.drop(['Unnamed: 0','PassengerId', 'Group', 'Group_size', 'Age_group', 'Cabin_number'], axis=1, inplace=True)
```

## Log transform
- Die Logarithmentransformation wird verwendet, um die Schiefe in Verteilungen zu verringern, insbesondere bei großen Ausreißern. Sie kann es den Algorithmen erleichtern, die richtigen Beziehungen zu "lernen". Wir werden sie auf die Ausgabenmerkmale anwenden, da diese durch Ausreißer stark verzerrt sind.

```python
# Plot log transform results
fig=plt.figure(figsize=(12,20))
for i, col in enumerate(['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Expenditure']):
    plt.subplot(6,2,2*i+1)
    sns.histplot(X[col], binwidth=100)
    plt.ylim([0,200])
    plt.title(f'{col} (original)')
    
    plt.subplot(6,2,2*i+2)
    sns.histplot(np.log(1+X[col]), color='C1')
    plt.ylim([0,200])
    plt.title(f'{col} (log-transform)')
    
fig.tight_layout()
plt.show()
```

```python
# Apply log transform
for col in ['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Expenditure']:
    X[col]=np.log(1+X[col])
    X_test[col]=np.log(1+X_test[col])
```

```python
X.head()
```

##  Encoding and scaling

```python
# 1. Indentify numerical and categorical columns
numerical_cols = [cname for cname in X.columns if X[cname].dtype in ['int64', 'float64']]
categorical_cols = [cname for cname in X.columns if X[cname].dtype == "object"]

# 2. Scale numerical data to have mean=0 and variance=1
numerical_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# 3. One-hot encode categorical data
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(drop='if_binary', handle_unknown='ignore',sparse=False))])

# 4. Combine preprocessing
ct = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)],
        remainder='passthrough')

# 5. Apply preprocessing
X = ct.fit_transform(X)
X_test = ct.transform(X_test)
```

## Erläuterung
1. Identifizierung numerischer und kategorischer Spalten:
Hier werden die numerischen und kategorischen Spalten im Datensatz getrennt. Der Code überprüft den Datentyp jeder Spalte im Dataframe X und kategorisiert sie in die Listen numerical_cols oder categorical_cols. Numerische Spalten haben die Datentypen 'int64' oder 'float64', während kategoriale Spalten den Datentyp 'object' haben.
2. Skalierung numerischer Daten:
Der Code definiert einen numerical_transformer-Pipeline mit sklearn.preprocessing.StandardScaler(). Der StandardScaler ist eine Vorverarbeitungsklasse, die numerische Daten skaliert, indem sie den Mittelwert subtrahiert und durch die Standardabweichung teilt. Die Pipeline ermöglicht die Kombination mehrerer Vorverarbeitungsschritte, obwohl in diesem Fall nur der Scaler enthalten ist.
3. One-Hot-Encoding kategorischer Daten:
Der Code definiert eine categorical_transformer-Pipeline mit sklearn.preprocessing.OneHotEncoder(). Der OneHotEncoder ist eine Vorverarbeitungsklasse, die kategoriale Daten in eine binäre One-Hot-Encoding-Darstellung umwandelt. Er weist jeder Kategorie in einer Spalte einen binären Wert (0 oder 1) zu. Die Pipeline enthält den OneHotEncoder mit zusätzlichen Parametern, um eine der binären Spalten für binäre Merkmale zu entfernen (drop='if_binary') und unbekannte Kategorien bei der Codierung zu ignorieren (handle_unknown='ignore').
4. Kombination der Vorverarbeitungsschritte:
Der Code erstellt ein ColumnTransformer-Objekt namens ct. Dieser Transformer ermöglicht das Anwenden verschiedener Vorverarbeitungsschritte auf verschiedene Spalten des Datensatzes. Er nimmt eine Liste von Transformern entgegen, wobei jeder Transformer als Tupel aus drei Elementen definiert ist: der Name des Transformers ('num' oder 'cat' in diesem Fall), das Transformer-Objekt (numerical_transformer oder categorical_transformer) und die Liste der Spalten, auf die der Transformer angewendet werden soll (z. B. numerical_cols oder categorical_cols). Der Parameter remainder='passthrough' gibt an, dass alle verbleibenden Spalten, die nicht explizit transformiert werden, unverändert weitergegeben werden sollen.
5. Anwendung der Vorverarbeitung:
Der Code wendet die im ColumnTransformer definierten Vorverarbeitungsschritte auf die Variable X mithilfe der Methode fit_transform() an. Dabei werden die Daten transformiert, indem die numerischen Spalten mit der numerical_transformer-Pipeline skaliert und die kategorischen Spalten mit der categorical_transformer-Pipeline in One-Hot-Encoding umgewandelt werden. Die transformierten Daten werden wieder in X gespeichert. Zusätzlich werden die X_test-Daten mit der Methode transform() des ColumnTransformer transformiert. Dadurch werden dieselben Transformationen wie bei den X-Daten angewendet, um eine konsistente Vorverarbeitung sicherzustellen.

```python
# Print new shape
print('Training set shape:', X.shape)
print('Test set shape:', X_test.shape)
```

```python
# Train-validation split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,stratify=y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
print(f"X_train shape = {X_train.shape}")
print(f"y_train shape = {y_train.shape}")
print(f"X_valid shape = {X_valid.shape}")
print(f"y_valid shape = {y_valid.shape}")
```

# 7. Modeling

Hier wird ein dictionary erstellt mit den 2 Algorithmen Logistic Regression und Support Vectore Machine. Zusätzlich werden parameter für den grid search festgelegt, um diese anschließend mit grid search zu trainieren

```python
# Classifiers
classifiers = {
    "LogisticRegression" : LogisticRegression(random_state=0),
    "SVC" : SVC(random_state=0, probability=True),
}

# Grids for grid search
LR_grid = {'penalty': ['l1','l2'],
          'C': [0.25, 0.5, 0.75, 1, 1.25, 1.5],
           'max_iter': [50, 100, 150]}


SVC_grid = {'C': [0.25, 0.5, 0.75, 1, 1.25, 1.5],
            'kernel': ['linear', "poly", 'rbf'],
            'gamma': ['scale', 'auto']}


# Dictionary of all grids
grid = {
    "LogisticRegression" : LR_grid,
    "SVC" : SVC_grid,
}
```

```python
# Train models with grid search

i=0
clf_best_params=classifiers.copy()
valid_scores=pd.DataFrame({'Classifer':classifiers.keys(), 'Validation accuracy': np.zeros(len(classifiers)), 'Training time': np.zeros(len(classifiers))})
for key, classifier in classifiers.items():
    start = time.time()
    clf = GridSearchCV(estimator=classifier, param_grid=grid[key], n_jobs=-1, cv=None)

    # Train and score
    clf.fit(X_train, y_train)
    valid_scores.iloc[i,1]=clf.score(X_valid, y_valid)
    
    # Save trained model
    clf_best_params[key]=clf.best_params_
    
    # Print iteration and training time
    stop = time.time()
    valid_scores.iloc[i,2]=np.round((stop - start)/60, 2)
    
    print('Model:', key)
    print('Training time (mins):', valid_scores.iloc[i,2])
    print('')
    i+=1
```

```python
# Show results
valid_scores
```

```python
# Show best parameters from grid search
clf_best_params
```

```python
# Classifiers
best_classifiers = {
    "LogisticRegression" : LogisticRegression(**clf_best_params["LogisticRegression"], random_state=0),
    "SVC" : SVC(**clf_best_params["SVC"], random_state=0, probability=True)
    }
```

## Confusion Matrix & F1 Score

```python
# Iterate over the classifiers in the best_classifiers dictionary
for key, classifier in best_classifiers.items():
    # Fit the classifier on the training data
    classifier.fit(X_train, y_train)

    # Predict on the validation data
    y_pred = classifier.predict(X_valid)

    # Compute the F1 score
    f1 = f1_score(y_valid, y_pred)
    print("F1 Score for", key, ":", f1)

    # Compute the confusion matrix
    c_mat = confusion_matrix(y_valid, y_pred)
    #print("Confusion Matrix for", key)
    #print(c_mat)

    # Display the confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=c_mat,
                                  display_labels=np.unique(y_valid))
    disp.plot()
    plt.title(key + " Confusion Matrix")
    plt.show()
```

# Confusion Matrix example to not get confused


## Cross validation and ensembling predictions

```python
# Number of folds in cross validation
FOLDS=10

preds=np.zeros(len(X_test))
for key, classifier in best_classifiers.items():
    start = time.time()
    
    # 10-fold cross validation
    cv = StratifiedKFold(n_splits=FOLDS, shuffle=True, random_state=0)
    
    score=0
    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y)):
        # Get training and validation sets
        X_train, X_valid = X[train_idx], X[val_idx]
        y_train, y_valid = y[train_idx], y[val_idx]

        # Train model
        clf = classifier
        clf.fit(X_train, y_train)

        # Make predictions and measure accuracy
        preds += clf.predict_proba(X_test)[:,1]
        score += clf.score(X_valid, y_valid)

    # Average accuracy    
    score=score/FOLDS
    
    # Stop timer
    stop = time.time()

    # Print accuracy and time
    print('Model:', key)
    print('Average validation accuracy:', np.round(100*score,2))
    print('Training time (mins):', np.round((stop - start)/60,2))
    print('')
    
# Ensemble predictions
preds=preds/(FOLDS*len(best_classifiers))
```

## Submission

```python
# Sample submission (to get right format)
sub=pd.read_csv('sample_submission.csv')

# convert the continuous prediction values to binary (True/False) values based on the threshold of 0.5
preds = [True if p >= 0.5 else False for p in preds]

# Add predictions
sub['Transported']=preds

# Prediction distribution
plt.figure(figsize=(6,6))
sub['Transported'].value_counts().plot.pie(explode=[0.1,0.1], autopct='%1.1f%%', shadow=True, textprops={'fontsize':16}).set_title("Prediction distribution")
```
